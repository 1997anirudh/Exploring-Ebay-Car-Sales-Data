```python
import pandas as pd
import numpy as np 
autos = pd.read_csv("autos.csv", encoding = "Latin-1")
autos.head()
autos.info()
```

```python
autos.columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'abtest',
       'vehicle_type', 'registration_year', 'gearbox', 'powerPS', 'model',
       'odometer', 'registration_month', 'fuel_type', 'brand',
       'unrepaired_damage', 'ad_created', 'nr_of_pictures', 'postal_code',
       'last_seen']
autos.head()
```

```python
autos["price"] = autos["price"].str.replace("$","").str.replace(",","").astype(float)
autos["odometer"] = autos["odometer"].str.replace("km","").str.replace("kms","").str.replace(",","").astype(float)
autos.rename({"odometer": "odometer_km"}, axis =1, inplace = True)
autos.rename({"price": "price_dollars"}, axis =1, inplace = True)

autos.head()
```

```python
brand_unique = autos["brand"].unique()
brand_perc = autos["brand"].value_counts(normalize = True )* 100
top_20_brands = brand_perc.head(20)
brands_over_5 = top_20_brands[top_20_brands > 5].index
```

```python
brands_over_5 = top_20_brands[top_20_brands > 5].index
brand_dict = { }
for brand in brands_over_5:
    ind_brand = autos[autos["brand"] == brand]
    one_brand_mean_cost = ind_brand["price_dollars"].mean()
    brand_dict[brand] = int(one_brand_mean_cost)
brand_dict
        
        
```

The above analysis is based of the brand of the car. Created a frwquency table to check which brand is present the most in the given dataframe. 
We only took only the brands that are present more than 5 percent in the given data. 
Those are the following 
1. audi
2. bmw
3. ford 
4. mercedes 
5. opel 
6. volkswagen 

We have found the average/mean price of these vehicles. I have created a dictionary to note them down in a presentable manner  

```python
mileage_dict = { }
for brand in brands_over_5:
    ind_mileage = autos[autos["brand"] == brand]
    one_mileage_mean_cost = ind_mileage["odometer_km"].mean()
    mileage_dict[brand] = int(one_mileage_mean_cost)
mileage_dict
        
        
```

```python
brand_series = pd.Series(brand_dict).sort_values(ascending = True)
mileage_series = pd.Series(mileage_dict)
brand_mileage_dataframe = pd.DataFrame(brand_series, columns=['mean_price'])
brand_mileage_dataframe["mean_mileage"] = mileage_series
brand_mileage_dataframe
```

```python
def freq_table(brand_name):
    brand_models = autos[autos["brand"] == brand_name]
    frequency_model = { }
    for val in brand_models["model"]:
        if val in frequency_model:
            frequency_model[val] += 1
        else:
            frequency_model[val] = 1
    return frequency_model
freq_table("audi")
```

A function pf freq_table has been created which takes 1 variable, which is the brand's name of the car. It will return the frequancy table of the various models in that particular brand. 

This will help us discern further which is the model under a particular brand that has the most models. 
Given an example of Audi above
